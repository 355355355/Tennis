```python
import cv2
import IPython.display as display
import PIL.Image

# 打开视频文件
video_path = "output_videos/output_video1.avi"
cap = cv2.VideoCapture(video_path)

# 检查视频是否成功打开
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# 读取并显示视频帧
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 转换帧为 PIL 图像并显示
    img = PIL.Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    display.display(img)
    display.clear_output(wait=True)
# 释放视频捕获对象
cap.release()
```
